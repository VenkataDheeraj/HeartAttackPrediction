```python
dqr = {}
for col in continuous_features:
  dqr[col] = {
      'Feature': col,
      'Desc.': 'Continuous',
      'Count': heart_df[col].count(),
      '% of Missing': 100*heart_df[col].isna().sum()/heart_df.shape[0],
      'Card.': heart_df[col].nunique(),
      'Min.': heart_df[col].min(),
      'Q1': heart_df[col].quantile(0.25),
      'Median': heart_df[col].median(),
      'Q3': round(heart_df[col].quantile(0.75),2),
      'Max.': heart_df[col].max(),
      'Mean': round(heart_df[col].mean(),2),
      'Std. Dev.': round(heart_df[col].std(),2)
  }
continuous_dqr = pd.DataFrame(dqr).T
print('Continuous Data Quality Report')
continuous_dqr
```

```python
dqr = {}
cat_cols = ['gender', 'class']
for col in cat_cols:
  dqr[col] = {
      'Feature': col,
      'Desc.': 'Categorical',
      'Count': heart_df[col].count(),
      '% of Missing': 100*heart_df[col].isna().sum()/heart_df.shape[0],
      'Card.': heart_df[col].nunique(),
      'Mode': heart_df[col].mode()[0],
      'Mode Freq.': heart_df[col].value_counts()[heart_df[col].mode()[0]],
      'Mode %': round((heart_df[col].value_counts()[heart_df[col].mode()[0]]*100)/heart_df.shape[0],2),
      '2nd Mode': heart_df[col].value_counts().index[1],
      '2nd Mode Freq.': heart_df[col].value_counts()[heart_df[col].value_counts().index[1]],
      '2nd Mode Perc': round((heart_df[col].value_counts()[heart_df[col].value_counts().index[1]]*100)/heart_df.shape[0],2)
      }
categorical_dqr = pd.DataFrame(dqr).T
print('Categorical Data Quality Report')
categorical_dqr
```

```python
import math

sns.set_theme(style="whitegrid", palette="muted")
continuous_color = "dodgerblue"
categorical_palette = ["#4E79A7", "#F28E2C"]

num_continuous = len(continuous_features)

cols = 3
rows = math.ceil(num_continuous / cols)

fig, axes = plt.subplots(rows, cols, figsize=(15, 5 * rows))
axes = axes.flatten()

for i, col in enumerate(continuous_features):
    sns.histplot(data=heart_df, x=col, kde=True, bins=30, color=continuous_color, edgecolor="black", alpha=0.8, ax=axes[i])
    axes[i].set_title(f"Distribution of {col}", fontsize=14)
    axes[i].set_xlabel(col, fontsize=12)
    axes[i].set_ylabel("Frequency", fontsize=12)

for j in range(num_continuous, len(axes)):
    axes[j].axis('off')
plt.suptitle('Histograms for continuous features', fontsize = 16, weight='bold')
plt.tight_layout()
plt.show()

fig, axes = plt.subplots(1, len(categorical_features), figsize=(12, 6))
axes = axes.flatten()

for i, col in enumerate(categorical_features):
    sns.countplot(
        data=heart_df, 
        x=col, 
        hue=col,
        palette=categorical_palette, 
        edgecolor="black", 
        alpha=0.9, 
        ax=axes[i]
    )
    axes[i].set_title(f"Count of {col}", fontsize=14)
    axes[i].set_xlabel(col, fontsize=12)
    axes[i].set_ylabel("Count", fontsize=12)
    if axes[i].get_legend() is not None:
        axes[i].get_legend().remove()
plt.suptitle('Barplots for categorical features', fontsize = 16, weight='bold')
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(4, 2, figsize=(14, 20))
axes = axes.flatten()

for i, col in enumerate(continuous_features):
    sns.boxplot(data=heart_df, x=col, ax=axes[i])
    axes[i].set_title(f'Boxplot of {col}', fontsize=14)
    axes[i].set_xlabel(col, fontsize=12)

plt.suptitle('Boxplots for continuous features', fontsize = 16, weight = 'bold')
axes[-1].axis('off')
plt.tight_layout()
plt.show()
```

Analysis from the above boxplots:
1. age:
   there are outliers, but no need to remove the outliers as we are not looking for specific age range.
2. impulse:
   there is an outlier in data which says impulse is more than 1000, which is error in measurement and as for other outliers, values above 200 bpm can be considered error measurement or a very rare case.
3. pressure height:
   outliers can be a valid case of heart attack, hence wont be making changes to outliers, but will be capping the values to remove extreme outliers
4. pressure low:
   there are some outliers, would be analysed as that kind of values mean the patient is higly likely to have an heart attack.
5. glucose:
   many outliers present, but that is expected if the patient is diabetic. will be capping extreme values
6. kcm:
   cannot remove any outliers but will be capping them, presence of kcm means there is heart damage, higher the kcm higher the damage, outliers will be capped to remove exteme values adn will be removed if they are erroneous
7. troponin:
   high outlier, cannot remove them as high troponin is significant for diagnosing heart attacks
   both troponin and kcm will be analysed together to find any erroneous instances
   

```python
def cap_outliers(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    data[column] = data[column].clip(lower=lower_bound, upper=upper_bound)
    return data

retain_columns = ['kcm', 'troponin']
cap_columns = ['impluse', 'pressurehight', 'pressurelow', 'glucose']
for col in cap_columns:
    dataset = cap_outliers(heart_df, col)

log_transform_columns = ['glucose', 'kcm', 'troponin']
for col in log_transform_columns:
    dataset[col] = np.log1p(heart_df[col])

for col in cap_columns + log_transform_columns:
    print(f"Updated range for {col}: {heart_df[col].min()} to {heart_df[col].max()}")
```

```python
from sklearn.preprocessing import StandardScaler

features_to_scale = ['age', 'impluse', 'pressurehight', 'pressurelow']
log_normalized_features = ['glucose', 'kcm', 'troponin']
scaler = StandardScaler()

scaled_features = scaler.fit_transform(heart_df[features_to_scale])
scaled_df = pd.DataFrame(scaled_features, columns=features_to_scale)
log_normalized_df = heart_df[log_normalized_features]
final_df = pd.concat([scaled_df, log_normalized_df, heart_df[['gender', 'class']]], axis=1)

final_df['bp_ratio'] = heart_df['pressurehight'] / heart_df['pressurelow']

final_df['heart_rate_category'] = pd.cut(
    heart_df['impluse'], bins=[0, 60, 100, 200], labels=['Low', 'Normal', 'High']
)

final_df['age_group'] = pd.cut(
    heart_df['age'], bins=[0, 40, 60, 100], labels=['Young', 'Middle-Aged', 'Elderly']
)
final_df['pressure_glucose_interaction'] = heart_df['pressurehight'] * heart_df['glucose']
```
